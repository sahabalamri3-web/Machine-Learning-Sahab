```python
# Data points (students marks)
data = [35, 40, 42, 60, 65, 70]

# Number of clusters
K = 2

# Initial centroids
centroids = [35, 70]

while True:
    # Create empty clusters
    cluster1 = []
    cluster2 = []

    # Assign points to nearest centroid
    for point in data:
        d1 = abs(point - centroids[0])
        d2 = abs(point - centroids[1])

        if d1 <= d2:
            cluster1.append(point)
        else:
            cluster2.append(point)

    # Calculate new centroids
    new_c1 = sum(cluster1) / len(cluster1)
    new_c2 = sum(cluster2) / len(cluster2)

    # Stop if centroids do not change
    if new_c1 == centroids[0] and new_c2 == centroids[1]:
        break

    centroids[0] = new_c1
    centroids[1] = new_c2

# Final result
print("Cluster 1:", cluster1)
print("Cluster 2:", cluster2)
print("Centroids:", centroids)
```
